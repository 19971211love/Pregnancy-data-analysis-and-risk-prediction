```python
import json
```

```python
data_dict = []
```

```python
for i in range(34):
    with open("../static/geometryProvince/"+str(i+1)+".json", "r",encoding='utf-8') as rf:
        js=json.load(rf)
        feat = js['features']
        data_dict.append([item["properties"]['name'] for item in feat])
```

```python
data_dict
```

```python
s='{\nitems:\n[\n'
for i in range(len(data_dict)):
        s+=('  {\n  \"id\": '+str(i+1)+',\n  \"counties\":\n  [\n')
        for j in range(len(data_dict[i])):
            s+=("    {\"name\":\" "+data_dict[i][j] +"\", \"value\":  100}")
            if j!=len(data_dict[i])-1:
                s+=(",\n")
        s += ('\n  ]\n  }')
        if i!=len(data_dict)-1:
            s += ','
        s += '\n'
s += (']\n}\n')
s
```

```python
with open("data.json", "wb") as wf:
    wf.write(s.encode('utf-8'))
```
